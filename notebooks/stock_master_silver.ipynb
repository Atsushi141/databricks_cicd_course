```python
from pyspark.sql import functions as F
```

```python
df = spark.table("`yfinance-dev`.bronze.stock_master")
```

```python
stock_master_df = (df.filter(F.col("市場・商品区分") != "ETF・ETN")
    .withColumnRenamed("日付", "date")
    .withColumnRenamed("コード", "code")
    .withColumnRenamed("銘柄名", "stock_name")
    .withColumnRenamed("市場・商品区分", "market_category")
    .withColumnRenamed("33業種コード", "industry_code_33")
    .withColumnRenamed("33業種区分", "industry_category_33")
    .withColumnRenamed("17業種コード", "industry_code_17")
    .withColumnRenamed("17業種区分", "industry_category_17")
    .withColumnRenamed("規模コード", "scale_code")
    .withColumnRenamed("規模区分", "scale_category")
)
```

```python
stock_master_df.write.mode("overwrite").saveAsTable("`yfinance-dev`.silver.stock_master")
```
